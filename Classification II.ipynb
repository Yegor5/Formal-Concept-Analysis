## Устойчивость к шуму

#### Заметим, что текущий алгоритм неустойчив к шуму. Тогда попробуем усовершенствовать его, используя метод RULEARNER. В моем случае некоторые объекты обучающей выборки будут рассматриваться как ненадежные. Тогда, сформивовав решетку формальных понятий, обозначим неблагонадежными все понятия, объекты которых содержат только один объект, и он является ненадежным, остальные понятия будем считать благонадежными. Далее будем подниматься по решетке понятий, и для каждой вершины будем считать, сколько неблаго-/благонадежных вершин в их нижнем конусе (верхним конусом вершины называется множество вершин, включая данную, которые могут быть достигнуты от данной вершины, поднимаясь только по восходящим ребрам, нижним конусом вершины называется множество вершин, которые содержат данную вершину в своем верхнем конусе). Тогда, если процент неблагонадженых вершин не превышает некоторого заранее заданного параметра, то и сама эта вершина является благонадежной, в противном случае, неблагонадежной. Неблагонадежные понятия не участвуют в формировании гипотез. Таким образом удается понизить влияние шума на формирование гипотез и на качество работы алгоритма

```python
def rulearner(lattice, graph, broken, percent):
    lattice_h = [1] * len(lattice)
    for i in range(1, len(lattice)):
        if len(lattice[i][0]) == 1:
            if lattice[i][0][0] in broken:
                lattice_h[i] = 0
            else:
                lattice_h[i] = 1
        else:
            truesum = 0
            falsesum = 0
            peaks = set()
            peaks.add(i)
            for edge in graph[::-1]:
                if edge[1] in peaks and edge[0] not in peaks:
                    peaks.add(edge[0])
                    if lattice_h[edge[0]]:
                        truesum += 1
                    else:
                        falsesum += 1
            if 100 * falsesum / (truesum + falsesum) > percent:
                lattice_h[i] = 0
    return lattice_h
```

## Временная сложность 
#### Так как операция построения алгебраической решетки требует пройтись циклом по всем возможным пересечениям объектов, то это уже дает экспоненциальную сложность алгоритму. Также для каждого пересечения объектов мы должны построить вектор признаков, объединяя векторы признаков всех объектов, которые входят в исходное пересечение. Помимо этого, для правильного построения решетки, для каждого вектора признаков необходимо найти наибольшее возможное пересечение объектов, которым присущ данный вектор признаков. Обе эти операции дают еще множитель n в оценке сложности. В итоге получаем, что временная сложность алгоритма $O(n * 2^{n})$. Проведем исследование: протестируем алгоритм построения решетки на разных данных и посмотрим, сколько времени необходимо, чтобы он успешно выполнился

Такая временная сложность делает невозможным применение алгоритма для решения задач с большим количеством данных.
